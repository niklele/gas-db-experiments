```python
def google_geocoding(stations_df, skip):
    locations = {} # station_id to google location object
    failures = {} # station_id to address string
    for station_id,data in stations_df.iterrows():
        if station_id in skip:
            continue
        print("{0},\"{1}\"".format(station_id, data.address))
        try:
            loc = geolocator.geocode(data.address)
        except Exception as e:
            print(e)
            failures[station_id] = data.address
        else:
            locations[station_id] = loc
        time.sleep(3) # rate limiting
    return locations, failures
```

```python
def successive_google_geocoding(stations_df, base_locations):
    locations = base_locations.copy()
    while len(locations) < 336:
        temp_locations, temp_failures = google_geocoding(stations_df, locations.keys())
        print(len(temp_locations))
        print(len(temp_failures))
        locations.update(temp_locations)
    return locations
```
